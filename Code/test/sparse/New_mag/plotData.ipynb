```python
import matplotlib.pyplot as plt

methods = ['Review Paper', 'PCA+SVM', 'Sparse+DBC', 'Sparse+SVM']
accuracy = [61.25, 70.75, 52, 84.25]

# Create the bar chart
plt.figure(figsize=(8, 6))
plt.bar(methods, accuracy)

# Add labels and title
plt.xlabel('Method Name', fontsize=12)
plt.ylabel('Classification Accuracy (%)', fontsize=12)
plt.title('Classification Accuracy Comparison', fontsize=14)

# Display the chart
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for the bar chart
methods = ['Review Paper', 'PCA+SVM', 'Sparse+DBC', 'Sparse+SVM']
accuracy = [61.25, 70.75, 52, 84.25]
missed_detection_rate = [0.35, 0.38, 0.84, 0.19]
false_alarm_rate = [0.43, 0.21, 0.13, 0.13]
authentication_rate = [0.61, 0.71, 0.52, 0.84]

# Create the bar chart for Classification Accuracy
plt.figure(figsize=(8, 6))
plt.bar(methods, accuracy, color='blue')

# Add labels and title for Classification Accuracy
plt.xlabel('Method Name', fontsize=12)
plt.ylabel('Classification Accuracy (%)', fontsize=12)
plt.title('Classification Accuracy Comparison', fontsize=14)

# Display the chart for Classification Accuracy
plt.tight_layout()
plt.show()

# Create the bar chart for Missed Detection Rate
plt.figure(figsize=(8, 6))
plt.bar(methods, missed_detection_rate, color='orange')

# Add labels and title for Missed Detection Rate
plt.xlabel('Method Name', fontsize=12)
plt.ylabel('Missed Detection Rate', fontsize=12)
plt.title('Missed Detection Rate Comparison', fontsize=14)

# Display the chart for Missed Detection Rate
plt.tight_layout()
plt.show()

# Create the bar chart for False Alarm Rate
plt.figure(figsize=(8, 6))
plt.bar(methods, false_alarm_rate, color='green')

# Add labels and title for False Alarm Rate
plt.xlabel('Method Name', fontsize=12)
plt.ylabel('False Alarm Rate', fontsize=12)
plt.title('False Alarm Rate Comparison', fontsize=14)

# Display the chart for False Alarm Rate
plt.tight_layout()
plt.show()

# Create the bar chart for Authentication Rate
plt.figure(figsize=(8, 6))
plt.bar(methods, authentication_rate, color='purple')

# Add labels and title for Authentication Rate
plt.xlabel('Method Name', fontsize=12)
plt.ylabel('Authentication Rate', fontsize=12)
plt.title('Authentication Rate Comparison', fontsize=14)

# Display the chart for Authentication Rate
plt.tight_layout()
plt.show()
```
