```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Define scenarios and classification methods
scenarios = ["Symmetric", "Symmetric Variable Speed", "Asymmetric", "Asymmetric Reflector"]
methods = ["Method 1", "Method 2", "Method 3"]

# Hypothetical Confusion Matrix values for 3 methods in each scenario
confusion_matrices = {
    "Symmetric": {
        "Method 1": [[85, 10], [5, 80]],
        "Method 2": [[80, 15], [10, 75]],
        "Method 3": [[78, 20], [12, 70]],
    },
    "Symmetric Variable Speed": {
        "Method 1": [[80, 15], [10, 75]],
        "Method 2": [[78, 20], [12, 70]],
        "Method 3": [[75, 25], [15, 65]],
    },
}

# Create heatmaps for each scenario-method pair
fig, axes = plt.subplots(len(confusion_matrices), len(methods), figsize=(12, 6))

for row, (scenario, methods_data) in enumerate(confusion_matrices.items()):
    for col, (method, matrix) in enumerate(methods_data.items()):
        matrix = np.array(matrix)
        sns.heatmap(matrix, annot=True, fmt="d", cmap="coolwarm",
                    xticklabels=["Predicted Positive", "Predicted Negative"],
                    yticklabels=["Actual Positive", "Actual Negative"],
                    ax=axes[row, col])
        axes[row, col].set_title(f"{scenario} - {method}")

# Show the plots
plt.tight_layout()
plt.show()
```
