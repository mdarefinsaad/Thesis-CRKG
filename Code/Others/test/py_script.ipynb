```python
t = np.linspace(0, 1, 500)  # time vector
print(t)
y_t = np.exp(-t) * (np.cos(2 * np.pi * 10 * t) + 1j * np.sin(2 * np.pi * 10 * t))  # example CIR
print(y_t)

# Compute the absolute value of y(t)
abs_y_t = np.abs(y_t)
print(abs_y_t)

# Plotting
plt.figure(figsize=(10, 4))
plt.plot(t, abs_y_t, label='|y(t)|')
plt.title('Absolute Value of the Channel Impulse Response')
plt.xlabel('Time (s)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.legend()
plt.show()
```
