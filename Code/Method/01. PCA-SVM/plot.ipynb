```python
# method 1 - PCA + SVM
import numpy as np
import matplotlib.pyplot as plt

# Define components and scenarios
components = [2, 3, 4, 5, 6]
scenarios = ["Symmetric", "Symmetric VarSpeed", "Asymmetric", "Asymmetric Reflector"]

# Hypothetical accuracy values for each scenario in each component
accuracy_data = np.array([
    [65, 68, 69, 70, 71],  # Symmetric
    [60, 59, 58, 60, 64],  # Symmetric VarSpeed
    [66, 67, 69, 75, 76],  # Asymmetric
    [61, 63, 67, 70, 71],  # Asymmetric Reflector
])

# Bar width and X-axis positions
bar_width = 0.15
x = np.arange(len(components))

# Define custom colors for each scenario
colors = ["#1f77b4", "#6fd1ff", "#2ca02c", "#00da6a"]  # Blue, Orange, Green, Red

# Create the bar chart
fig, ax = plt.subplots(figsize=(10, 6))
for i, (scenario, color) in enumerate(zip(scenarios, colors)):
    ax.bar(x + i * bar_width, accuracy_data[i], width=bar_width, label=scenario, color=color)

# Formatting
ax.set_xlabel("Number of components", fontsize=12)
ax.set_ylabel("Accuracy (%)", fontsize=12)
ax.set_title("Method PCA-SVM", fontsize=14)
ax.set_xticks(x + bar_width * (len(scenarios) / 2 - 0.5))
ax.set_xticklabels(components, fontsize=11)
# ax.legend(fontsize=8)
ax.grid(axis="y", linestyle="--", alpha=0.7)
ax.legend(fontsize=8, title="Scenarios", loc='upper left', bbox_to_anchor=(1, 1))  # Moves legend outside
ax.set_ylim(0, 90)
# ax.set_ylim(0, 100)
# Show the plot
plt.show()
```
