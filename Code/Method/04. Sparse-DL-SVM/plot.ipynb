```python
# Method 4 - Sparse-DL-SVM
import numpy as np
import matplotlib.pyplot as plt

# Define components and scenarios
no_atoms = [5, 10, 15, 20]  # Corrected variable name
scenarios = ["Symmetric", "Symmetric VarSpeed", "Asymmetric", "Asymmetric Reflector"]

# Hypothetical accuracy values for each scenario in each component
accuracy_data = np.array([
    [74, 78, 77, 76],  # Symmetric
    [68, 73, 74, 73],  # Symmetric VarSpeed
    [81, 84, 84, 85],  # Asymmetric
    [74, 75, 76, 78],  # Asymmetric Reflector
])

# Bar width and X-axis positions
bar_width = 0.15
x = np.arange(len(no_atoms))  # Fixed variable name

# Define custom colors for each scenario
colors = ["#1f77b4", "#6fd1ff", "#2ca02c", "#00da6a"]  # Blue, Light Blue, Green, Light Green

# Create the bar chart
fig, ax = plt.subplots(figsize=(10, 6))
for i, (scenario, color) in enumerate(zip(scenarios, colors)):
    ax.bar(x + i * bar_width, accuracy_data[i], width=bar_width, label=scenario, color=color)

# Formatting
ax.set_xlabel("Number of atoms", fontsize=12)
ax.set_ylabel("Accuracy (%)", fontsize=12)
ax.set_title("Method Sparse-DL-SVM", fontsize=14)
ax.set_xticks(x + bar_width * (len(scenarios) / 2 - 0.5))
ax.set_xticklabels(no_atoms, fontsize=11)
ax.legend(fontsize=8, title="Scenarios", loc='upper left', bbox_to_anchor=(1, 1))  # Moves legend outside
ax.grid(axis="y", linestyle="--", alpha=0.7)
ax.set_ylim(0, 90)

# Show the plot
plt.show()
```
