```python
# Method 2 - SRC
import numpy as np
import matplotlib.pyplot as plt

# Define components and scenarios
non_zero_coeff = [2, 4, 8, 10]  # Corrected variable name
scenarios = ["Symmetric", "Symmetric VarSpeed", "Asymmetric", "Asymmetric Reflector"]

# Hypothetical accuracy values for each scenario in each component
accuracy_data = np.array([
    [59, 65, 58, 62],  # Symmetric
    [58, 60, 56, 54],  # Symmetric VarSpeed
    [61, 62, 60, 58],  # Asymmetric
    [65, 62, 60, 61],  # Asymmetric Reflector
])

# Bar width and X-axis positions
bar_width = 0.15
x = np.arange(len(non_zero_coeff))  # Fixed avariable name

# Define custom colors for each scenario
colors = ["#1f77b4", "#6fd1ff", "#2ca02c", "#00da6a"]  # Blue, Light Blue, Green, Light Green

# Create the bar chart
fig, ax = plt.subplots(figsize=(10, 6))
for i, (scenario, color) in enumerate(zip(scenarios, colors)):
    ax.bar(x + i * bar_width, accuracy_data[i], width=bar_width, label=scenario, color=color)

# Formatting
ax.set_xlabel("Non-Zero Coefficients", fontsize=12)
ax.set_ylabel("Accuracy (%)", fontsize=12)
ax.set_title("Method - Sparse Residual Classifier", fontsize=14)
ax.set_xticks(x + bar_width * (len(scenarios) / 2 - 0.5))
ax.set_xticklabels(non_zero_coeff, fontsize=11)
ax.legend(fontsize=11, title="Scenarios", loc='upper left', bbox_to_anchor=(1, 1))  # Moves legend outside
ax.grid(axis="y", linestyle="--", alpha=0.7)

ax.set_ylim(0, 90)
# Show the plot
plt.show()
```
